```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
def convert_sqft_to_num(x):
    piece= x.split('-')
    if len(piece)==2:
        return (float(piece[0])+float(piece[1]))/2
    try:
        return float(x)
    except:
        return None
    
```

```python
def remove_pps_outliers(df):
    df_out= pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m= np.mean(subdf.price_per_sqft)
        st= np.std(subdf.price_per_sqft)
        reduced_df= subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out= pd.concat([df_out, reduced_df],ignore_index=True)
    return df_out
```

```python
df5= df4.copy()
df5 = remove_pps_outliers(df4)
df5.shape
```

```python
X= df10.drop(['price'],axis='columns')
X.head()
```

```python
Y= df10.price
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25, random_state=10)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
model= LinearRegression()
model.fit(X_train,Y_train)
model.score(X_test,Y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
cv= ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
cross_val_score(LinearRegression(), X,Y, cv=cv)
```

```python
# After performing 5 fold cross validation on Linear Regression, maximun time the result is above 80%. 
# So we take Linear Regression as our appropiate model for house prediction system.

def predict_price(location, sqft,bath,bhk):
    loc_index = np.where(X.columns==location)[0][0]
    
    x= np.zeros(len(X.columns))
    x[0] =  sqft
    x[1]= bath
    x[2]= bhk
    if loc_index >=0:
      x[loc_index]=1
    return model.predict ([x])[0]
```

```python
predict_price('1st Block Jayanagar',1000, 2, 2)
```

```python
predict_price('Vishwapriya Layout',1000, 2, 3)
```
